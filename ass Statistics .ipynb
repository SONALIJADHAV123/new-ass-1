1. Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list. After generating the list, find the following:
(i) Write a Python function to calculate the mean of a given list of numbers.
Create a function to find the median of a list of numbers.
(ii) Develop a program to compute the mode of a list of integers.
(iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.
(iv) Write a Python function to find the geometric mean of a list of positive numbers.
(v) Create a program to calculate the harmonic mean of a list of values.
(vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).
(vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.

```python
import random
import statistics
import numpy as np
from scipy.stats import gmean, hmean

# Generate a list of 100 integers containing values between 90 to 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Function to calculate the mean
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

# Function to find the median
def calculate_median(numbers):
    numbers.sort()
    n = len(numbers)
    mid = n // 2
    if n % 2 == 0:
        return (numbers[mid - 1] + numbers[mid]) / 2
    else:
        return numbers[mid]

# Function to compute the mode
def calculate_mode(numbers):
    return statistics.mode(numbers)

# Function to calculate the weighted mean
def calculate_weighted_mean(values, weights):
    weighted_sum = sum(v * w for v, w in zip(values, weights))
    total_weight = sum(weights)
    return weighted_sum / total_weight

# Function to calculate the geometric mean
def calculate_geometric_mean(numbers):
    return gmean(numbers)

# Function to calculate the harmonic mean
def calculate_harmonic_mean(numbers):
    return hmean(numbers)

# Function to determine the midrange
def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

# Function to find the trimmed mean
def calculate_trimmed_mean(numbers, trim_percent):
    n = len(numbers)
    k = int(n * trim_percent / 100)
    trimmed_numbers = sorted(numbers)[k:-k]
    return calculate_mean(trimmed_numbers)

# Example usage
print("List:", int_list)
print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", calculate_mode(int_list))
print("Weighted Mean:", calculate_weighted_mean(int_list, [1] * 100))  # Example with equal weights
print("Geometric Mean:", calculate_geometric_mean(int_list))
print("Harmonic Mean:", calculate_harmonic_mean(int_list))
print("Midrange:", calculate_midrange(int_list))
print("Trimmed Mean (10%):", calculate_trimmed_mean(int_list, 10))
```

(iii) Create a program to find the variance and standard deviation of a list of numbers.
(iv) Implement a function to compute the interquartile range (IQR) of a list of values.
(v) Build a program to calculate the coefficient of variation for a dataset.
(vi) Write a Python function to find the mean absolute deviation (MAD) of a list of numbers.
(vii) Create a program to calculate the quartile deviation of a list of values.
(viii) Implement a function to find the range-based coefficient of dispersion for a dataset

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Generate a list of 500 integers containing values between 200 to 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Function to calculate the range
def calculate_range(numbers):
    return max(numbers) - min(numbers)

# Function to calculate variance and standard deviation
def calculate_variance(numbers):
    return np.var(numbers)

def calculate_std_dev(numbers):
    return np.std(numbers)

# Function to compute the interquartile range (IQR)
def calculate_iqr(numbers):
    q75, q25 = np.percentile(numbers, [75 ,25])
    return q75 - q25

# Function to calculate the coefficient of variation
def calculate_cv(numbers):
    mean = np.mean(numbers)
    std_dev = np.std(numbers)
    return std_dev / mean

# Function to calculate the mean absolute deviation (MAD)
def calculate_mad(numbers):
    mean = np.mean(numbers)
    return np.mean(np.abs(numbers - mean))

# Function to calculate the quartile deviation
def calculate_quartile_deviation(numbers):
    q75, q25 = np.percentile(numbers, [75 ,25])
    return (q75 - q25) / 2

# Function to find the range-based coefficient of dispersion
def calculate_range_based_cd(numbers):
    range_val = calculate_range(numbers)
    mean = np.mean(numbers)
    return range_val / mean

# Visualization of Frequency & Gaussian distribution
def plot_frequency_gaussian(numbers):
    plt.figure(figsize=(10, 6))
    sns.histplot(numbers, kde=False, color='blue', bins=20, label='Frequency')
    mu, std = norm.fit(numbers)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p * len(numbers) * (xmax - xmin) / 20, 'k', linewidth=2, label='Gaussian distribution')
    plt.legend()
    plt.title('Frequency & Gaussian Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()

# Visualization of Frequency smoothened KDE plot
def plot_frequency_kde(numbers):
    plt.figure(figsize=(10, 6))
    sns.histplot(numbers, kde=True, color='blue', bins=20)
    plt.title('Frequency Smoothened KDE Plot')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()

# Visualization of Gaussian distribution & smoothened KDE plot
def plot_gaussian_kde(numbers):
    plt.figure(figsize=(10, 6))
    sns.kdeplot(numbers, color='blue', label='KDE')
    mu, std = norm.fit(numbers)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2, label='Gaussian distribution')
    plt.legend()
    plt.title('Gaussian Distribution & Smoothened KDE Plot')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()

# Example usage
print("List:", int_list2)
print("Range:", calculate_range(int_list2))
print("Variance:", calculate_variance(int_list2))
print("Standard Deviation:", calculate_std_dev(int_list2))
print("Interquartile Range (IQR):", calculate_iqr(int_list2))
print("Coefficient of Variation (CV):", calculate_cv(int_list2))
print("Mean Absolute Deviation (MAD):", calculate_mad(int_list2))
print("Quartile Deviation:", calculate_quartile_deviation(int_list2))
print("Range-based Coefficient of Dispersion:", calculate_range_based_cd(int_list2))

# Visualization
plot_frequency_gaussian(int_list2)
plot_frequency_kde(int_list2)
plot_gaussian_kde(int_list2)
```

3. Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

```python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities must have the same length.")
        if not (0.999 <= sum(probabilities) <= 1.001):  # Allowing a small margin for floating point arithmetic
            raise ValueError("Sum of probabilities must be 1.")
        self.values = values
        self.probabilities = probabilities
    
    def expected_value(self):
        
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))
    
    def variance(self):
       
        mean = self.expected_value()
        return sum(prob * (value - mean) ** 2 for value, prob in zip(self.values, self.probabilities))

# Example usage:
values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]

random_variable = DiscreteRandomVariable(values, probabilities)

print("Expected Value:", random_variable.expected_value())
print("Variance:", random_variable.variance())
```

4. Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.

```python
import random

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
       
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities must have the same length.")
        if not (0.999 <= sum(probabilities) <= 1.001):  # Allowing a small margin for floating point arithmetic
            raise ValueError("Sum of probabilities must be 1.")
        self.values = values
        self.probabilities = probabilities
    
    def expected_value(self):
        
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))
    
    def variance(self):
       
        mean = self.expected_value()
        return sum(prob * (value - mean) ** 2 for value, prob in zip(self.values, self.probabilities))

def simulate_die_rolls(n):
    
    outcomes = [0] * 6  # Initialize a list to count outcomes for each face

    for _ in range(n):
        roll = random.randint(1, 6)
        outcomes[roll - 1] += 1

    probabilities = [count / n for count in outcomes]
    return probabilities

# Number of simulations
n = 10000
values = [1, 2, 3, 4, 5, 6]
probabilities = simulate_die_rolls(n)

# Create an instance of DiscreteRandomVariable for the die
die = DiscreteRandomVariable(values, probabilities)

print("Expected Value:", die.expected_value())
print("Variance:", die.variance())
```

5. Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.

```python
import numpy as np

def generate_random_samples(distribution, params, sample_size):
   
    if distribution == 'binomial':
        if 'n' not in params or 'p' not in params:
            raise ValueError("Binomial distribution requires 'n' (number of trials) and 'p' (probability of success).")
        samples = np.random.binomial(n=params['n'], p=params['p'], size=sample_size)
    elif distribution == 'poisson':
        if 'lambda' not in params:
            raise ValueError("Poisson distribution requires 'lambda' (rate parameter).")
        samples = np.random.poisson(lam=params['lambda'], size=sample_size)
    else:
        raise ValueError("Unsupported distribution. Supported distributions are 'binomial' and 'poisson'.")
    
    mean = np.mean(samples)
    variance = np.var(samples)
    
    return {'mean': mean, 'variance': variance}

# Example usage:
binomial_params = {'n': 10, 'p': 0.5}
poisson_params = {'lambda': 4}

binomial_result = generate_random_samples('binomial', binomial_params, 10000)
poisson_result = generate_random_samples('poisson', poisson_params, 10000)

print("Binomial Distribution (n=10, p=0.5):", binomial_result)
print("Poisson Distribution (lambda=4):", poisson_result)
```

8. Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.

```python
import numpy as np

def normal_pdf(x, mu, sigma):
   
    exponent = -((x - mu) ** 2) / (2 * sigma ** 2)
    
    pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(exponent)
    
    return pdf

sigma = 1
x_values = np.linspace(-5, 5, 100)  # Generate x values for plotting
pdf_values = normal_pdf(x_values, mu, sigma)

print("PDF at x = 0:", normal_pdf(0, mu, sigma))

import matplotlib.pyplot as plt
plt.plot(x_values, pdf_values)
plt.title('Normal Distribution PDF')
plt.xlabel('x')
plt.ylabel('PDF')
plt.grid(True)
plt.show()
```

9. Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

```python
import numpy as np

def exponential_cdf(x, lam):
   
    cdf = 1 - np.exp(-lam * x)
    return cdf

# Example usage:
lam = 0.5  # Rate parameter (lambda) of the exponential distribution
x_values = np.linspace(0, 5, 100)  # Generate x values for plotting
cdf_values = exponential_cdf(x_values, lam)

# Print the CDF values at specific x points
print("CDF at x = 1:", exponential_cdf(1, lam))

# Plot the CDF
import matplotlib.pyplot as plt
plt.plot(x_values, cdf_values)
plt.title('Exponential Distribution CDF')
plt.xlabel('x')
plt.ylabel('CDF')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Data for old and new layouts
old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)  # 70 purchases out of 1000 visitors

# Number of successes (purchases) and trials (visitors) for old and new layouts
successes_old = np.sum(old_layout)
trials_old = len(old_layout)
successes_new = np.sum(new_layout)
trials_new = len(new_layout)

# Applying z-test
z_stat, p_value = proportions_ztest([successes_old, successes_new], [trials_old, trials_new])

# Output results
print("Z-statistic:", z_stat)
print("P-value:", p_value)

# Interpret the results
if p_value < 0.05:
    print("There is significant evidence to reject the null hypothesis.")
    if z_stat > 0:
        print("The new layout leads to a higher conversion rate.")
    else:
        print("The old layout leads to a higher conversion rate.")
else:
    print("There is not enough evidence to reject the null hypothesis. The layouts may have similar conversion rates.")
```

```python
import numpy as np
from scipy.stats import ttest_ind

def ab_test_analysis(data1, data2):
    
    t_statistic, p_value = ttest_ind(data1, data2)

    # Calculate degrees of freedom
    dof = len(data1) + len(data2) - 2

    return t_statistic, dof, p_value

# Data for click-through rates
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform A/B test analysis
t_statistic, dof, p_value = ab_test_analysis(layout_a_clicks, layout_b_clicks)

# Output results
print("t-statistic:", t_statistic)
print("Degrees of freedom:", dof)
print("p-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is significant evidence that the two layouts have different click-through rates.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the two layouts have different click-through rates.")
```
