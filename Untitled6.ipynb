```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)  # Generating 1000 random numbers from a normal distribution

# Visualize the distribution of the numerical variable using seaborn
plt.figure(figsize=(8, 6))
sns.histplot(data, kde=True, color='skyblue', edgecolor='black')  # Kernel Density Estimation (KDE) is enabled
plt.title('Distribution of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
import numpy as np

from bokeh.plotting import figure, show
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.transform import transform
import numpy as np

# Generate random data
np.random.seed(0)
n = 100
x = np.random.random(size=n) * 10  # Random x values
y = np.random.random(size=n) * 10  # Random y values
sizes = np.random.randint(5, 20, size=n)  # Random sizes for markers
colors = np.random.random(size=n)  # Random colors for markers

# Create a Bokeh figure
p = figure(title="Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add scatter plot with different marker sizes and colors
p.scatter(x, y, radius=sizes, fill_color=linear_cmap('colors', 'Viridis256', 0, 1), fill_alpha=0.6, line_color="green")

# Add color bar
color_mapper = linear_cmap(field_name='colors', palette='Viridis256', low=0, high=1)
color_bar = ColorBar(color_mapper=color_mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits counts ['Apples', 'Oranges', 'Bananas', 'Pears'] [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")

# Add vertical bars
p.vbar(x='fruits', top='counts', width=0.9, source=source, color="#FF5733")

# Set axis labels
p.xaxis.axis_label = "Fruits"
p.yaxis.axis_label = "Counts"

# Remove gridlines
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

# Show the plot
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram", background_fill_color="#f0f0f0")

# Add a quad glyph to create the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.5)

# Customize the plot
p.y_range.start = 0
p.xaxis.axis_label = 'Value'
p.yaxis.axis_label = 'Frequency'

# Show the plot
show(p)
```
